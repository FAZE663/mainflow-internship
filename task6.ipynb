```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/fasal/Downloads/IPL_2023_DATASET.csv")
```

```python
df.columns.values
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull()
```

```python
sns.heatmap(df.isnull())
```

```python
df['COST IN ₹ (CR.)']=df['COST IN ₹ (CR.)'].fillna(0)
df['Cost IN $ (000)']=df['Cost IN $ (000)'].fillna(0)
df['Base Price IN $']=df['Base Price IN $'].fillna(0)
df['Base Price IN ₹']=df['Base Price IN ₹'].fillna(0)
df['Base Price']=df['Base Price'].fillna(0)
```

```python
df.isnull()
```

```python
sns.heatmap(df.isnull(),cmap='summer')
```

```python
df['2022 Squad']=df['2022 Squad'].fillna('No Team')
```

```python
sns.heatmap(df.isnull(),cmap='summer')
```
