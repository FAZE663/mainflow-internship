```python
qstns=["1. How many people have heart disease and how doesnt have heart disease?",
       "2. Poeple of which sex have most heart disease?",
       "3. People of which sex has which type of heart pain the most?",
       "4. People with which chest pain are more prone to heart disease?"]
```

```python
#answer to qs no 1

df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['green','yellow'])
plt.title("Heart disease values")
plt.xlabel("1=Heart disease, 2=No heart disease")
plt.ylabel("Amount")
```

```python
#pie chart
df.target.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Disease','No Disease'])

```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie')
plt.legend(['Male','Female'])
```

```python
#answer to 2nd qs
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title("Heart disease frequency for sex")
plt.xlabel("0=No Heart Disease, 1=Heart Disease")
```

```python
# qs 3
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar')
plt.title("chest pain type vs count")
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['orchid','green','blue','yellow'])
plt.title("type of chest pain for sex")
plt.xlabel('0 = female  1 = male')
```
