```python
data1 = data.fillna(value=0)
data1
```

```python
import numpy as np
from scipy import stats
```

```python
data1.columns
```

```python
data1.drop(['Observation'],axis=1,inplace=True)
data1.columns
```

```python
q1=data1.quantile(.25)
q2=data1.quantile(.75)
iqr=q2-q1
```

```python
iqr
```

```python
data1=data1[~((data1<(q1-1.5*iqr))|(data1>(q2+1.5*iqr))).any(axis=1)]
data1
```
